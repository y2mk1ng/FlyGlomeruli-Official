```python
import os, sys, cv2, numpy, pandas, csv
from matplotlib import pyplot as plt
from PIL import Image
from PIL import ImageColor
from collections import Counter
```

```python
# make the grayscale images binary with cv2
def binary_image(filename, file_out_binary):
    im_gray = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    # convert to binary: using cv2.threshold(<image>, <threshold value>, <max value>, <the style of thresholding>)
    ret, im_binary = cv2.threshold(im_gray, 1, 255, cv2.THRESH_BINARY)
    cv2.imwrite(file_out_binary, im_binary)
```

```python
# read the binary images as arrays 
def as_numpy_arr(file_out_binary, csv_file):
    img_bin = Image.open(file_out_binary)
    np_arr = numpy.array(img_bin)
    #print(np_arr)
    pandas.DataFrame(np_arr).to_csv(csv_file)    
```
